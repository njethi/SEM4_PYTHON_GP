```python
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
pll = parallel_coordinates(df,'Country', cols=['Latitude','Longitude','Average_Temperature (°C)','Humidity (%)','Wind_Speed (km/h)','Precipitation (mm)','Cloud_Cover (%)','UV_Index (mW/m2)','Air_Pressure (hPa)'] , color=('red','blue','green','orange','yellow','pink','violet','skyblue','black'))
plt.figure(figsize=(15,15))
plt.show()
```

```python
plt.subplot(1,4,1)
plt.boxplot(df['Average_Temperature (°C)'])
plt.xlabel("Average_Temperature (°C)'")
plt.subplot(1,4,2)
plt.boxplot(df["Humidity (%)"])
plt.xlabel("Humidity (%)")
plt.subplot(1,4,3)
plt.boxplot(df["Wind_Speed (km/h)"])
plt.xlabel("Wind_Speed (km/h)")
plt.subplot(1,4,4)
plt.boxplot(df["Precipitation (mm)"])
plt.xlabel("Precipitation (mm)s")
plt.show()
```

```python
from pywaffle import Waffle
fig = plt.figure(FigureClass = Waffle,rows = 150,values = df['Humidity (%)'],labels = list(df['Country']))
plt.figure(figsize=(15,15))
plt.show()
```

```python
import seaborn as sns
sns.heatmap(dd,annot=False,cmap='viridis',linewidths=0.5,linecolor='white')
plt.figure(figsize=(12,8))
plt.show()
```

```python
import folium
sanfran_map = folium.Map()
```

```python
incidents = folium.map.FeatureGroup()
# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df['Latitude'], df['Longitude']):
    incidents.add_child(folium.features.CircleMarker([lat, lng],radius=5,color='yellow',fill=True,fill_color='blue',fill_opacity=0.6))
sanfran_map.add_child(incidents)
sanfran_map
```
